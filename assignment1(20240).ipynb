```python
def func(x,y,w,k=6,z=None):
    if(z==None):
        return x+y
    elif(k==6):
        return x+y+k
    else:
        return x+y+w
print(func(2,3,4,5,6))
print(func(2,5,9))
print(func(2,6,4,7,10),"\n\n")

def func1(x):
    for i in range(0,6):
        while(x!=None):
            if(x<=10):
                print(x)
                x=x+1
                continue
            else:
                break
print(func1(1),"\n")
print(func1(6))
```

```python
arr = np.array([1,1,4,5,5,5,7])
q = {}
for i in arr:
  if i in q:
    q[i]+=1
  else:
    q[i]=1
arr1 = []
for i in q:
  if q[i]==1:
    arr1.append(i)
print("array : ",arr1)
```
